# analisis dataset titanic

tenggelamnya kapal titanic pada tahun 1912 merupakan salah satu tragedi maritim paling terkenal dalam sejarah. banyak penumpang tidak dapat diselamatkan karena keterbatasan sekoci. dataset ini digunakan untuk mempelajari pola keselamatan penumpang berdasarkan berbagai atribut seperti jenis kelamin usia kelas tiket dan faktor sosial ekonomi.

```python
# import library yang diperlukan
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
# memuat dataset dari github
base_url = 'https://raw.githubusercontent.com/aulyashabrina666-boop/partisipatif_ml_ds_1/f515da2c4798e507003f96c121e384eebb593860/data/'

train_df = pd.read_csv(base_url + 'train.csv')
test_df = pd.read_csv(base_url + 'test.csv')
gender_submission_df = pd.read_csv(base_url + 'gender_submission.csv')
```

```python
# menampilkan beberapa baris pertama dari dataset train
print('dataset train')
print(train_df.head())
print('\ninfo dataset train')
print(train_df.info())
print('\nringkasan statistik')
print(train_df.describe())

print('\n\ndataset test')
print(test_df.head())

print('\n\ndataset gender submission')
print(gender_submission_df.head())
```

# struktur dataset titanic

dataset titanic memiliki beberapa atribut utama yang digunakan dalam analisis:

atribut nominal:
* sex: jenis kelamin penumpang
* embarked: pelabuhan keberangkatan

atribut ordinal:
* pclass: kelas penumpang (1 untuk kelas atas, 2 untuk kelas menengah, 3 untuk kelas bawah)

atribut numerik:
* age: usia penumpang
* fare: harga tiket
* sibsp: jumlah saudara kandung atau pasangan di kapal
* parch: jumlah orang tua atau anak di kapal

atribut target:
* survived: status keselamatan penumpang (0 untuk tidak selamat, 1 untuk selamat)

```python
# preprocessing data
train_processed = train_df.copy()

# menghapus kolom yang tidak relevan
train_processed = train_processed.drop(['Name', 'Ticket', 'Cabin'], axis=1)

# menangani missing value pada kolom age dengan nilai rata rata
train_processed['Age'].fillna(train_processed['Age'].mean(), inplace=True)

# menangani missing value pada kolom embarked dengan nilai yang paling sering muncul
train_processed['Embarked'].fillna(train_processed['Embarked'].mode()[0], inplace=True)

# encoding atribut kategori sex
le_sex = LabelEncoder()
train_processed['Sex'] = le_sex.fit_transform(train_processed['Sex'])

# encoding atribut kategori embarked
le_embarked = LabelEncoder()
train_processed['Embarked'] = le_embarked.fit_transform(train_processed['Embarked'])

print('dataset setelah preprocessing')
print(train_processed.head())
print('\ninfo dataset setelah preprocessing')
print(train_processed.info())
```

# normalisasi data

normalisasi diperlukan agar atribut numerik memiliki skala yang sebanding sehingga model dapat belajar dari data secara lebih stabil. proses normalisasi mengubah nilai atribut ke dalam rentang tertentu seperti 0 sampai 1 sehingga tidak ada atribut yang mendominasi proses pembelajaran model machine learning.

```python
# normalisasi atribut numerik menggunakan min max scaling
train_normalized = train_processed.copy()

# memilih kolom numerik untuk dinormalisasi
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

scaler = MinMaxScaler()
train_normalized[numeric_cols] = scaler.fit_transform(train_normalized[numeric_cols])

print('dataset setelah normalisasi')
print(train_normalized.head(10))
print('\nringkasan statistik setelah normalisasi')
print(train_normalized[numeric_cols].describe())
```
